This is used to Identify pattern and Biase in the GenZ Data set.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("../data/GenZ_DatingApp_Data.csv")

#Checking the data types
print(data.dtypes)

# Converting the object data types to Categorys
for col in data.select_dtypes(include='object').columns:
    data[col] = data[col].astype('category')

# Converting the Satisfication column to category

data["Satisfaction"] = data["Satisfaction"].astype('category')
```

Question 1: Is there gender bias in dating app usage?

```python

# Count of each gender
gender_app_counts = data.groupby("Gender")["Primary_App"].value_counts().unstack()

# Plot bar chart
gender_app_counts.plot(kind="bar", stacked=True, figsize=(8, 5), colormap="viridis")

plt.xlabel("Gender")
plt.ylabel("Number of Users")
plt.title("Dating App Preferences by Gender")
plt.legend(title="Primary App")
plt.xticks(rotation=0)
plt.show()
```

Answer: Yes, there is a gender biase on the dating app being used.

Question 2: Are certain demographics overrepresented in the dataset?

```python
# Age Distribution
plt.figure(figsize=(8, 5))
sns.histplot(data['Age'], bins=10, kde=True, color="blue")
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution in Dataset")
plt.show()

# Chart2
plt.figure(figsize=(8, 5))
sns.kdeplot(data=data, x="Age", hue="Gender", fill=True, palette="viridis", alpha=0.5)

plt.xlabel("Age")
plt.ylabel("Density")
plt.title("Age Density Distribution by Gender")
plt.legend(title="Gender")
plt.show()


# Chart 3
plt.figure(figsize=(8, 5))
sns.histplot(data, x="Location", hue="Gender", bins=15, kde=False, palette="magma", alpha=0.7)

plt.xlabel("Location")
plt.ylabel("Count")
plt.title("Daily App Usage Distribution by Location")
plt.legend(title="Gender")
plt.show()

```

Answer: Yes, based on the age distribution, location, it could be seen that a certain demography are overrepresented and underrepresented in the data set
