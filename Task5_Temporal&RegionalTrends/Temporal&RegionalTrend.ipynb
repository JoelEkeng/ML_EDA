This is used to Identify pattern and Biase in the GenZ Data set.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("../data/GenZ_DatingApp_Data.csv")

# Converting the object data types to Categorys
for col in data.select_dtypes(include='object').columns:
    data[col] = data[col].astype('category')

# Converting the Satisfication column to category

data["Satisfaction"] = data["Satisfaction"].astype('category')
#Checking the data types
print(data.dtypes)
print(data.describe())
print("Age data:")
print(data[data['Age']>22].count())
```

Question 1: Are younger or older Gen-Z users more active on dating apps?


```python
# Plot with Seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(data=data, x="Daily_Usage_Time", y="Age",  marker="o", color="g")

# Labels & Title
plt.xlabel("Daily Usage Time")
plt.ylabel("Age")
plt.title("Most Active Over Time")
plt.xticks(rotation=45)
plt.yticks(ticks=np.arange(18, 26, 2))

plt.show()


# Using Usage Frequency

plt.figure(figsize=(8, 5))
sns.lineplot(data=data, x="Usage_Frequency", y="Age",  marker="o", color="g")

# Labels & Title
plt.xlabel("Usage Frequency")
plt.ylabel("Age")
plt.title("Most Active Over Time based on Usage Frequency")
plt.xticks(rotation=45)
plt.yticks(ticks=np.arange(18, 26, 2))

plt.show()

```

Answer: Yes, the younger GenZ are more active on the dating.

The median age is 22 and it could be seen that more GenZ less that 22 are more active on the dating app

Question 2: Does dating app preference change over time?

```python

# Bubble Chart
plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x="Age", y="Primary_App", size="Age", hue="Age", palette="coolwarm", sizes=(50, 500))

# Labels & Title
plt.xlabel("Age Group")
plt.ylabel("Dating App")
plt.title("Popularity of Dating Apps by Age Group")
plt.grid()

plt.show()
```

Answer: Yes,preferences change with age.
